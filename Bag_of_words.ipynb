## Bag of Words Implementations

```python
import pandas as pd
messages = pd.read_csv("krishnaik06 Complete-Data-Science-With-Machine-Learning-And-NLP-2024 main 26-CompleteNLP For Machine Learning-Practicals_SpamClassifier-master\smsspamcollection\SMSSpamCollection", sep="\t", names=["label", "message"])
```

```python
messages.head(11)
```

```python
## Data Cleaning and Preprocessing
import re
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()
```

```python
corpus = []
for i in range(len(messages)):
    words = re.sub("[^a-zA-Z]", " ", messages['message'][i])
    words = words.lower()
    words = words.split()
    words = [p_stemmer.stem(word) for word in words if word not in stopwords.words('english')]
    words = " ".join(words)
    corpus.append(words)
```

```python
corpus
```

```python
## creating the bag of words model
from sklearn.feature_extraction.text import CountVectorizer

## To apply Binary Bag of Words 
## use binary = True in CountVectorizer

cv = CountVectorizer(max_features = 2000) ## max_features allow us to select top "n" (2000 here) number of most frequent words from the entire corpus
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
## output feature
y = pd.get_dummies(messages['label'])
```

```python
y = y.iloc[:, 1].values
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect = MultinomialNB().fit(X_train, y_train)
```

```python
spam_detect
```

```python
## making the predictions
y_pred = spam_detect.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
