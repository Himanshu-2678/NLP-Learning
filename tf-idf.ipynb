### TF-IDF Implementation

```python
import pandas as pd
messages = pd.read_csv("krishnaik06 Complete-Data-Science-With-Machine-Learning-And-NLP-2024 main 26-CompleteNLP For Machine Learning-Practicals_SpamClassifier-master\smsspamcollection\SMSSpamCollection", sep="\t", names=["label", "message"])
```

```python
messages.head()
```

```python
## Data Cleaning and Preprocessing
import re ## re module is used for regular expression operations
import nltk
nltk.download("stopwords") 
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(len((messages))):
    words = re.sub("[^a-zA-Z]", " ", messages['message'][i])
    ## the above line replaces all the characters except a-z and A-Z with space
    words = words.lower()
    words = words.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')]
    words = " ".join(words)
    corpus.append(words)
```

```python
## implementing TF-IDF 
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features = 2000)
```

```python
## traning the model
X = vectorizer.fit_transform(corpus).toarray()
X
```
